```python
cols = ['Name','Gender','Count']
names_1900 = pd.read_csv('/Users/ashvets/Temp/names/yob1900.txt', names=cols)
names_1950 = pd.read_csv('/Users/ashvets/Temp/names/yob1950.txt', names=cols)
names_2000 = pd.read_csv('/Users/ashvets/Temp/names/yob2000.txt', names=cols)
names_1900_1950 = pd.merge(names_1900, names_1950, on=['Name', 'Gender'], suffixes=('_1900', '_1950'))
```

```python
names_1900_1950_2000 = pd.merge(names_1900_1950, names_2000, on=['Name', 'Gender'])
```

```python
def agg_count(row):
    row.Count = row.Count_1900 + row.Count_1950 + row.Count
    return row

names_1900_1950_2000.apply(agg_count, axis=1).sort_values(by='Count', ascending=False)
```

```python
names_by_year = {}
for year in range(1900, 2000, 10):
    names_by_year[year] = pd.read_csv(
        '/Users/ashvets/Temp/names/yob{}.txt'.format(year),
        names=['Name','Gender','Count']
    )
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])
names_all
```

```python
name_dynamics = names_all.groupby([names_all.index.get_level_values(0), 'Name']).sum()
```

```python
gender_dynamics = names_all.groupby([names_all.index.get_level_values(0), 'Gender']).sum()
```

```python
name_dynamics.query('Name == ["John", "Mary", "William"]').unstack('Name').plot()
```

```python
gender_dynamics_cols = gender_dynamics.unstack('Gender')
gender_dynamics_cols
```

```python
gender_dynamics_cols.plot()
```

```python
name_dynamics.query('Name == ["John", "Mary", "William"]').unstack('Name').plot.bar()
```

```python
names_all.groupby('Name').sum().sort_values(by='Count', ascending=False).head(5).plot.pie(y='Count')
```

```python
def name_to_len(row):
    row.Name = len(row.Name)
    return row
name_by_len = names.apply(name_to_len, axis=1)
name_by_len
```

```python
name_by_len.plot.scatter(x='Name', y='Count')
```

```python
def scatter_len(group):
    plt.plot(group['Name'], group['Count'], 'o', label=group.name)
    
name_by_len.groupby('Gender').apply(scatter_len)
plt.xlabel('Name')
plt.ylabel('Count')
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.charts import Bar
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_notebook()

p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

p.line(x, y, legend="Temp.", line_width=2)

show(p)
```

```python
name_common_dynamics = names_all.groupby(level=0).sum()
name_common_dynamics.reset_index(level=0, inplace=True)
name_common_dynamics
```

```python
p = Bar(name_common_dynamics, 'Year', values='Count')

show(p)
```
